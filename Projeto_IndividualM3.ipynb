##*Importando Bibliotecas*

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Dados dos Relatórios de Progresso

```python
url = 'https://raw.githubusercontent.com/guimoraes021/ProjetoIndividualM3/main/Relatorio1.csv'
pd.read_csv(url, sep=';')
dados = pd.read_csv(url, sep=';')
dados
```

## Criando DataFrame

```python
data = {
    "Dia": [0, 1, 2, 3, 4, 5, 6],
    "Horas Trabalhadas": [6, 7, 8, 6, 7, 5, 4],
    "Bugs Corrigidos": [3, 2, 1, 4, 3, 2, 1],
    "Tarefas Concluídas": [5, 4, 6, 4, 5, 3, 2]
}

df = pd.DataFrame(data)
```

### Total de Horas Trabalhadas

```python
total_horas_trabalhadas = df['Horas Trabalhadas'].sum()


plt.figure(figsize=(12,5))
plt.bar(dados['Dia'], dados['Horas Trabalhadas'], color='skyblue')
plt.title(f'Total de Horas Trabalhadas na Semana: {total_horas_trabalhadas} Horas')
plt.xlabel('Dia da Semana')
plt.ylabel('Horas Trabalhadas')


plt.show()
```

### Média Diária de Horas Trabalhadas

```python
media_horas_trabalhadas = df["Horas Trabalhadas"].mean()
media_horas_trabalhadas
print(f"Média Diária de Horas Trabalhadas: {media_horas_trabalhadas:.2f}")
```

### Total de Bugs Corrigidos

```python
total_bugs_corrigidos = df['Bugs Corrigidos'].sum()


plt.figure(figsize=(12, 5))
plt.bar(dados['Dia'], dados['Bugs Corrigidos'], color='green')
plt.title(f'Total de Bugs Corrigidos na Semana: {total_bugs_corrigidos} Bugs')
plt.xlabel('Dia da Semana')
plt.ylabel('Bugs Corrigidos')


plt.show()
```

### Média Diária de Bugs Corrigidos

```python
media_bugs_corrigidos = df["Bugs Corrigidos"].mean()
media_bugs_corrigidos
print(f"Média Diária de Horas Trabalhadas: {media_bugs_corrigidos :.2f}")
```

### Total de Tarefas Concluídas

```python
total_tarefas_concluidas = df['Tarefas Concluídas'].sum()


plt.figure(figsize=(12,5))
plt.bar(dados['Dia'], dados['Tarefas Concluídas'], color='black')
plt.title(f'Total de tarefas concluidas na Semana: {total_tarefas_concluidas} Tarefas Concluídas')
plt.xlabel('Dia da Semana')
plt.ylabel('Tarefas Concluídas')

plt.show()
```

### Média Diária de Tarefas Concluídas

```python
media_tarefas_concluidas = df["Tarefas Concluídas"].mean()
media_tarefas_concluidas
print(f"Média Diária de Tarefas Concluídas: {media_tarefas_concluidas:.2f}")
```

### Produtividade Diária (Tarefas concluídas por hora)

```python
plt.figure(figsize=(12, 5))
sns.lineplot(x="Dia", y="Produtividade Diária", data=df, marker="o")
plt.title("Produtividade Diária (Tarefas Concluídas por Hora)")
plt.xlabel("Dia da Semana")

plt.ylabel("Produtividade Diária")
plt.grid(True)


plt.figtext(0.5, 0.01, "Gráfico de Linha mostrando a Produtividade Diária", ha="center")

plt.show()
```
